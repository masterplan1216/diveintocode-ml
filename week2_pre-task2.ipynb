【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\begin{bmatrix}
1.0 & 0.8 \\\ 
0.8 & 1.0
\end{bmatrix}
$$

多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。  
作成したndarrayのshapeは(500, 2)になります。


《シード値について》  
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。

```python
import numpy as np
np.random.seed(0) #乱数のシード値を固定

mu = [-3, 0] #平均
sigma = [[1.0, 0.8], [0.8, 1.0]] #共分散行列

#2次元正規乱数を500個作成
random_ndarray = np.random.multivariate_normal(mu, sigma, 500)
print(random_ndarray)
print(random_ndarray.shape)
```

**【問題2】散布図による可視化**  
問題1で作成したデータ点を散布図により可視化してください。


散布図はmatplotlibのplt.scatter()を使うことで描けます。  
[matplotlib.pyplot.scatter — Matplotlib 3.1.1 documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter)

```python
from matplotlib import pyplot as plt

#散布図の作成
plt.scatter(random_ndarray[:, 0], random_ndarray[:, 1])
plt.title("random_ndarray scatter plot", fontsize=16)
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

**【問題3】ヒストグラムによる可視化**  
問題1で作成したデータをヒストグラムにより可視化してください。


ヒストグラムはplt.hist()を使うことで描けます。  
[matplotlib.pyplot.hist — Matplotlib 3.1.1 documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html)


各次元ごとに作成し、2つのヒストグラムとしてください。

**【問題4】データの追加**  
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\begin{bmatrix}
1.0 & 0.8 \\\ 
0.8 & 1.0
\end{bmatrix}
$$

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
#新たに2次元正規分布による乱数500個作成
np.random.seed(0)
mu2 = [0, -3]
sigma2 = [[1.0, 0.8], [0.8, 1.0]]
random_ndarray2 = np.random.multivariate_normal(mu2, sigma2, 500)
random_ndarray2

#散布図の表示
plt.scatter(random_ndarray[:, 0], random_ndarray[:, 1], label="0")
plt.scatter(random_ndarray2[:, 0], random_ndarray2[:, 1], label="1")
plt.title("random_ndarray&random_ndarray2 scatter plot", fontsize=16)
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.legend() #凡例の表示
plt.show()
```

**【問題5】データの結合**  
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。  
[numpy.concatenate — NumPy v1.16 Manual](https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html)


[numpy.vstack — NumPy v1.19 Manual](https://numpy.org/doc/stable/reference/generated/numpy.vstack.html)

```python
#np.concatenate()を使用
linking_ndarray = np.concatenate((random_ndarray, random_ndarray2), axis=0)

np.set_printoptions(threshold=3000) #printoptionを変更して省略をなくした
print(linking_ndarray)
print(linking_ndarray.shape)
```

```python
#np.vstack()を使用
linking2_ndarray = np.vstack((random_ndarray, random_ndarray2))
print(linking2_ndarray)
print(linking2_ndarray.shape)
```

**【問題6】ラベル付け**  
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
#np.concatenate()を使用
#次元を合わせることに留意
label_ndarray = np.hstack((np.zeros(500), np.ones(500)))
label_linking_ndarray = np.concatenate((linking_ndarray, label_ndarray.reshape(1000, 1)), axis=1)

np.set_printoptions(precision=5, suppress=True) #小数点以下5桁、指数表記無し
print(label_linking_ndarray)
```

```python
#np.append()を使用
#次元を合わせることに留意
label_ndarray = np.hstack((np.zeros(500), np.ones(500)))
label_linking_ndarray = np.append(linking_ndarray, label_ndarray.reshape(1000, 1), axis=1)

np.set_printoptions(precision=5, suppress=True)
print(label_linking_ndarray)
```

```python
#np.insert()を使用
#次元を合わせなくてもできた
label_ndarray = np.hstack((np.zeros(500), np.ones(500)))

label_linking_ndarray = np.insert(linking_ndarray, 2, label_ndarray, axis=1)

np.set_printoptions(precision=5, suppress=True)
print(label_linking_ndarray)
```
